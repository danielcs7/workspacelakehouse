```python

def config_env(env_path, key_name, key_value, overwrite=False) -> bool:
    """
    Args:
        env_path (str): Caminho da pasta e nome do arquivo
        key_name (str): Nome da chave
        key_value (str): O token da chave
        overwrite (bool): True to update a Key

    Returns:
        bool: Retorna True indicando operação bem sucedida, False para  erro.
    """
    # from dotenv import load_dotenv
    # import os

    try:
        if os.path.exists(env_path):
            print(f"Arquivo '{env_path}' já existe.")            

            # Carrega variáveis existentes
            load_dotenv(env_path)
            existing_key = os.getenv(key_name)        

            # Lê todas as linhas do .env
            with open(env_path, "r") as f:
                lines = f.readlines()

            with open(env_path, "w") as f:
                key_written = False
                for line in lines:
                    if line.strip().startswith(f"{key_name}="):
                        if overwrite:
                            f.write(f"{key_name}={key_value}\n")
                            print(f"Chave '{key_name}' atualizada no arquivo .env.")
                        else:
                            f.write(line)
                            print(f"Chave '{key_name}' já existe. Use overwrite=True para atualizar.")
                        key_written = True
                    else:
                        f.write(line)

                if not key_written:
                    f.write(f"{key_name}={key_value}\n")
                    print(f"Chave '{key_name}' adicionada ao arquivo existente.")

        else:
            # Cria o arquivo com a nova chave
            with open(env_path, "w") as f:
                f.write(f"{key_name}={key_value}\n")
            print(f"Arquivo '{env_path}' criado com a chave '{key_name}'.")

        return True

    except Exception as e:
        print(f"Erro ao manipular o arquivo .env: {e}")
```
