### Listar arquivos de Metadados e arquivos de Dados com SparkSQL
 
 - Listar arquivos de dados
 - Listar arquivos de metadados (Manifests)
 - Listar referencias, conhecidas como snapshots da tabela
 - Listar snapshots
 - Listar partições

Documentação: https://iceberg.apache.org/docs/latest/spark-queries/

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .appName('dataincode') \
    .config("spark.jars", "/opt/spark/jars/iceberg-spark-runtime-3.5_2.12-1.6.0.jar") \
    .config("spark.sql.extensions", "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions") \
    .config("spark.sql.catalog.local", "org.apache.iceberg.spark.SparkCatalog") \
    .config("spark.sql.catalog.spark_catalog.type", "hive") \
    .config("spark.sql.catalog.local.warehouse", "s3a://datalake/iceberg") \
    .getOrCreate()

#Ajuste de log WARN log para ERROR
spark.sparkContext.setLogLevel("ERROR")
```

```python
# Listar arquivos de dados
spark.sql("SELECT * FROM iceberg.bronze.vendas.files;").toPandas()
```

```python
# Listar arquivos de METADADOS
spark.sql("SELECT * FROM iceberg.bronze.vendas.manifests;").toPandas()
```

```python
spark.sql("SELECT * FROM iceberg.bronze.vendas.refs").toPandas()
```

```python
## Os snapshots capituram os metadados de uma tabela Iceberg de um ponto especifico no tempo,
## eles são imutaveis e garantem a consistencia e confiabilidade dos dados e possibilitam o "Time Travel"

spark.sql("SELECT * FROM iceberg.bronze.vendas.snapshots;").toPandas()
```

```python
spark.sql("SELECT * FROM iceberg.bronze.vendas.partitions;").show()
```

### É possivel relacionar o historico da tabela com a tabela de snapshots

```python
spark.sql(""" 
    SELECT
    	h.made_current_at,
    	s.operation,
    	h.snapshot_id,
    	h.is_current_ancestor,
    	s.summary['spark.app.id']
    FROM iceberg.bronze.vendas.history h
    JOIN iceberg.bronze.vendas.snapshots s
      ON h.snapshot_id = s.snapshot_id
    ORDER BY made_current_at;"""
).toPandas()
```
